```python
import cv2
import dlib
from scipy.spatial import distance
from imutils import face_utils
import pygame
ad=cv2.VideoCapture(0)
face_shape=dlib.get_frontal_face_detector()
dlib_landmark=dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
(l_start,l_end)=face_utils.FACIAL_LANDMARKS_68_IDXS["left_eye"]
(r_start,r_end)=face_utils.FACIAL_LANDMARKS_68_IDXS["right_eye"]

#Initialize Pygame and load music
pygame.mixer.init()
pygame.mixer.music.load('sound.wav')

def eye_ratio(eye):
    A=distance.euclidean(eye[1],eye[5])
    B=distance.euclidean(eye[2],eye[4])
    C=distance.euclidean(eye[0],eye[3])
    EAR=(A+B)/(2.0*C)
    return EAR
thresh = 0.25
frame_check = 20
flag=0

while True:
    sc,img=ad.read()
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_shape(gray,0)
    for face in faces:
        landmark=dlib_landmark(gray,face)
        landmark=face_utils.shape_to_np(landmark)
        
        left_eye=landmark[l_start:l_end]
        right_eye=landmark[r_start:r_end]
        
        left_ear=eye_ratio(left_eye)
        right_ear=eye_ratio(right_eye)
        ear=(left_ear+right_ear)/2.0
        
        left_eye_hull=cv2.convexHull(left_eye)
        right_eye_hull=cv2.convexHull(right_eye)
        #cv2.drawContours(img,[left_eye_hull],-1,(0,0,255),1)
        #cv2.drawContours(img,[right_eye_hull],-1,(0,0,255),1)
        
        if ear < thresh:
            flag +=1
            #print(flag)
            if flag >= frame_check:
                pygame.mixer.music.play()
                pygame.mixer.music.set_volume(1.0)
                cv2.putText(img, "****************DROWSINESS DETECTED****************", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                cv2.putText(img, "****************ALERT!STOP!****************", (10,325),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                
        else:
            pygame.mixer.music.stop()
            flag=0
    cv2.imshow("Frame", img)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        cv2.destroyAllWindows()
        ad.release()
        break
         
    #cv2.imshow("Frame", img)
    #cv2.waitKey(0)
```
